```python
import numpy as np
import matplotlib.pyplot as plt
import one_over_f_analysis as oof
```

```python
list_2048 = ['2025_01_28_102027', '2025_01_28_103529', '2025_01_28_105031', '2025_01_28_110533', '2025_01_28_112035', '2025_01_28_140012',
             '2025_01_28_141514', '2025_01_28_143015']
list_4096 = ['2025_01_28_113537', '2025_01_28_115039', '2025_01_28_120540', '2025_01_28_122042', '2025_01_28_123544', '2025_01_28_144517',
             '2025_01_28_150019', '2025_01_28_151521']
```

```python
cw_avg_arr_2048 = []
cw_freqs_2048 = 0
for n in list_2048:
    cw_test = oof.OneOverFAnalyser('Data/CW_Test_2048/', n)
    cw_test.compute_psd_for_cw(save=True, savepath='2048_CW_Test_psd/', isolate_cw=True, average_to_time=True, average_to_time_s=1)
    cw_avg_arr_2048.append(cw_test.mean_psd)
    cw_freqs_2048 = cw_test.freqs_psd
    del cw_test

cw_avg_arr_2048 = np.array(cw_avg_arr_2048)
cw_avg_2048 = np.mean(cw_avg_arr_2048, axis=0)
cw_std_2048 = np.std(cw_avg_arr_2048, axis=0)
print(cw_avg_2048)
print(cw_freqs_2048)
```

```python

cw_avg_arr_4096 = []
cw_freqs_4096 = 0

for n in list_4096:
    cw_test = oof.OneOverFAnalyser('Data/CW_Test_4096/', n)
    cw_test.compute_psd_for_cw(save=True, savepath='4096_CW_Test_psd/', isolate_cw=True, average_to_time=True, average_to_time_s=1)
    print(cw_test.)
    cw_test.plot_psd()
    cw_avg_arr_4096.append(cw_test.mean_psd)
    cw_freqs_4096 = cw_test.freqs_psd
    del cw_test
cw_avg_arr_4096 = np.array(cw_avg_arr_4096)
cw_avg_4096 = np.mean(cw_avg_arr_4096, axis=0)
cw_std_4096 = np.std(cw_avg_arr_4096, axis=0)
print(cw_avg_4096)
print(cw_freqs_4096)

```

```python

plt.plot(cw_freqs_2048[1:], cw_avg_2048[1:], label = r'$\delta\nu = $'+str(int(2.048*10**6 / 2048)) + ' Hz')
plt.plot(cw_freqs_4096[1:], cw_avg_4096[1:], label = r'$\delta\nu = $'+str(int(2.048*10**6 / 4096)) + ' Hz')
plt.legend()
plt.ylabel('PSD')
plt.xlabel('Temporal Frequency [Hz]')
plt.xscale('linear')
plt.yscale('log')
plt.tight_layout()
plt.grid()
plt.show()

```

```python
cw70 = oof.OneOverFAnalyser(file_root='Data/', filename='2025_01_16_104232')
cw70.load_psd('SavedPSDs/')
cw65 = oof.OneOverFAnalyser(file_root='Data/', filename='2025_01_16_110340')
cw65.load_psd('SavedPSDs/')
cw75 = oof.OneOverFAnalyser(file_root='Data/', filename='2025_01_16_113134')
cw75.load_psd('SavedPSDs/')

arr = [cw65, cw70, cw75]
label = ['65 MHz', '70 MHz', '75 MHz']

for i in np.arange(len(arr)):
    plt.plot(arr[i].freqs_psd[1:], arr[i].mean_psd[1:], label=label[i])
plt.xlabel('Temporal Frequency [Hz]')
plt.ylabel('PSD')
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.tight_layout()
plt.show()

mean_psd = np.mean(np.array([cw65.mean_psd, cw70.mean_psd, cw75.mean_psd]), axis=0)
std_psd = np.mean(np.array([cw65.mean_psd, cw70.mean_psd, cw75.mean_psd]), axis=0)
freqs_psd = cw70.freqs_psd

plt.plot(freqs_psd[1:], mean_psd[1:], label='Averaged CW PSD')
plt.xlabel('Temporal Frequency [Hz]')
plt.ylabel('PSD')
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.tight_layout()
plt.show()

dummy = oof.OneOverFAnalyser(file_root='Data/', filename='egg')
dummy.mean_psd = mean_psd
dummy.std_psd = std_psd
dummy.freqs_psd = freqs_psd
dummy.fit_for_psd_params(no_plot=False, save_image=False)
```
